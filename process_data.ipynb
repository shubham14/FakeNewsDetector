```python
import torch
import pandas as pd
import numpy as np
import os
import pickle
import torchtext
```

```python
df = open(r'C:\\Users\\Shubham\\Desktop\\FakeNewsDetector\\data\\train.csv', encoding='utf8')
```

```python
text, labels = [], []

for i, line in enumerate(df):    
    if i == 0:
        continue
    t, l = line.split('\t')
    text.append(t)
    labels.append(l)
```

```python
labels = list(map(lambda x: ord(x[0]) - 48, labels))
```

```python
zipped = list(zip(text, labels))
```
